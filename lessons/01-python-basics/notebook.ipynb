# Antes de empezar: ¿Cómo funciona Google Colab?

Google Colab es un **cuaderno interactivo** que corre en tu navegador. Tiene dos tipos de celdas:

- **Celdas de texto** (como esta) — explican lo que vamos a hacer.
- **Celdas de código** (las grises) — aquí es donde se escribe y ejecuta código Python.

Vamos a practicar cómo usarlo antes de empezar con la lección.

### Paso 1: Ejecutar una celda

Para ejecutar una celda de código, haz clic en ella y luego:
- Presiona el botón de **play** (▶) que aparece a la izquierda, o
- Presiona **Shift + Enter** en tu teclado.

Pruébalo con la celda de abajo:

````python
print("¡Hola! Esta celda se ejecutó correctamente")
````

Si ves el texto `¡Hola! Esta celda se ejecutó correctamente` abajo de la celda, ¡felicitaciones! Ya sabes ejecutar código.

### Paso 2: Colab "recuerda" lo que ejecutaste

Cuando ejecutas una celda que crea una variable, Colab la guarda en memoria. Las celdas siguientes pueden usar esa variable sin tener que crearla de nuevo.

Ejecuta las dos celdas de código siguientes **en orden** para verlo en acción:

````python
# Celda A: creamos una variable
mensaje = "Colab me recuerda"
print(mensaje)
````

````python
# Celda B: usamos la variable creada en la celda anterior
print(mensaje + " desde otra celda")
````

La Celda B no creó la variable `mensaje`, pero pudo usarla porque Colab recuerda que ya la creaste en la Celda A.

### Paso 3: ¿Qué pasa si reinicio la sesión?

A veces las cosas dejan de funcionar y necesitas empezar de cero. Para eso existe **reiniciar la sesión**:

**Entorno de ejecución → Reiniciar sesión** (o `Ctrl + M` y luego `.`)

Al reiniciar, Colab **borra todas las variables de la memoria**. Es como si no hubieras ejecutado ninguna celda.

**Pruébalo ahora:**
1. Ve a **Entorno de ejecución → Reiniciar sesión**
2. Sin ejecutar la Celda A, intenta ejecutar solo la **Celda B** de arriba
3. Verás un error en rojo: `NameError: name 'mensaje' is not defined` — porque Colab ya no recuerda la variable

Después de reiniciar, simplemente ejecuta las celdas de nuevo desde arriba y todo volverá a funcionar.

## 1. Variables

Una **variable** es un nombre que le damos a un valor para poder usarlo después. Es como una caja con una etiqueta donde guardamos algo.

En Python, creamos una variable simplemente asignándole un valor con `=`. Usamos `print()` para mostrar el valor en pantalla.

Los tipos de datos básicos son:
- **`int`** — números enteros (ej: `5`, `100`)
- **`float`** — números con decimales (ej: `3.14`, `0.5`)
- **`str`** — texto (ej: `"hola"`)
- **`bool`** — verdadero o falso (`True`, `False`)

````python
# Ejemplo: crear variables de distintos tipos
edad = 45
altura = 1.75
nombre = "Ana"
es_estudiante = True

print(nombre)
print(edad)
print(altura)
print(es_estudiante)
````

## 2. Operadores Básicos

````python
# Ejemplo: operaciones básicas
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
````

## 3. Strings (Texto)

Los **strings** son cadenas de texto. Se escriben entre comillas (`"buenas"` o `'buenas'`).

Cosas útiles que podemos hacer con strings:
- **Unir** dos textos con `+` (concatenar)
- **Saber cuántas letras tiene** con `len()`
- **Convertir a mayúsculas** con `.upper()`
- **Convertir a minúsculas** con `.lower()`

````python
# Ejemplo: operaciones con strings
saludo = "Hola"
nombre = "María"

# Concatenar (unir) dos strings
mensaje = saludo + " " + nombre
print(mensaje)

# Largo del string
print(len(mensaje))

# Mayúsculas y minúsculas
print(mensaje.upper())
print(mensaje.lower())
````

## 4. Listas

Una **lista** es una colección ordenada de elementos. Se escribe entre corchetes `[]` y los elementos se separan con comas.

- Accedemos a un elemento por su **índice** (posición), que empieza en `0`.
- Usamos `len()` para saber cuántos elementos tiene.
- Usamos `.append()` para agregar un elemento al final.

````python
# Ejemplo: listas
frutas = ["manzana", "plátano", "naranja"]

# Acceder al primer elemento (índice 0)
print(frutas[0])

# Cuántos elementos tiene la lista
print(len(frutas))

# Agregar un elemento
frutas.append("uva")
print(frutas)
````

## 5. Condicionales

Los **condicionales** nos permiten ejecutar código solo si se cumple una condición.

```python
if condición:
    # se ejecuta si la condición es verdadera
else:
    # se ejecuta si la condición es falsa
```

**Importante:** el código dentro del `if` o `else` debe estar **indentado** (con 4 espacios o un tab).

````python
# Ejemplo: condicionales
temperatura = 35

if temperatura > 30:
    print("Hace mucho calor")
else:
    print("La temperatura está bien")
````

## 6. Ciclos (for)

Un **ciclo `for`** nos permite repetir una acción para cada elemento de una lista.

```python
for elemento in lista:
    # hacer algo con cada elemento
```

````python
# Ejemplo: recorrer una lista con for
animales = ["gato", "perro", "conejo"]

for animal in animales:
    print(animal)
````

## 7. Funciones

Una **función** es un bloque de código reutilizable. La definimos con `def`, le damos un nombre, y puede recibir **parámetros** y **devolver** un resultado con `return`.

```python
def nombre_funcion(parametro):
    # hacer algo
    return resultado
```

````python
# Ejemplo: definir y usar una función
def saludar(nombre):
    return "Hola, " + nombre

resultado = saludar("Carlos")
print(resultado)
````
