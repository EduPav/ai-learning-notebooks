# Antes de empezar: ¿Cómo funciona Google Colab?

Google Colab es un **cuaderno interactivo** que corre en tu navegador. Tiene dos tipos de celdas:

- **Celdas de texto** (como esta) — explican lo que vamos a hacer.
- **Celdas de código** (las grises) — aquí es donde se escribe y ejecuta código Python.

Vamos a practicar cómo usarlo antes de empezar con la lección.

### Paso 1: Ejecutar una celda

Para ejecutar una celda de código, haz clic en ella y luego:
- Presiona el botón de **play** (▶) que aparece a la izquierda, o
- Presiona **Shift + Enter** en tu teclado.

Pruébalo con la celda de abajo:

````python
print("¡Hola! Esta celda se ejecutó correctamente")
````

Si ves el texto `¡Hola! Esta celda se ejecutó correctamente` abajo de la celda, ¡felicitaciones! Ya sabes ejecutar código.

### Paso 2: Colab "recuerda" lo que ejecutaste

Cuando ejecutas una celda que crea una variable, Colab la guarda en memoria. Las celdas siguientes pueden usar esa variable sin tener que crearla de nuevo.

Ejecuta las dos celdas de código siguientes **en orden** para verlo en acción:

````python
# Celda A: creamos una variable
mensaje = "Colab me recuerda"
print(mensaje)
````

````python
# Celda B: usamos la variable creada en la celda anterior
print(mensaje + " desde otra celda")
````

La Celda B no creó la variable `mensaje`, pero pudo usarla porque Colab recuerda que ya la creaste en la Celda A.

### Paso 3: ¿Qué pasa si reinicio la sesión?

A veces las cosas dejan de funcionar y necesitas empezar de cero. Para eso existe **reiniciar la sesión**:

**Entorno de ejecución → Reiniciar sesión** (o `Ctrl + M` y luego `.`)

Al reiniciar, Colab **borra todas las variables de la memoria**. Es como si no hubieras ejecutado ninguna celda.

**Pruébalo ahora:**
1. Ve a **Entorno de ejecución → Reiniciar sesión**
2. Sin ejecutar la Celda A, intenta ejecutar solo la **Celda B** de arriba
3. Verás un error en rojo: `NameError: name 'mensaje' is not defined` — porque Colab ya no recuerda la variable

Después de reiniciar, simplemente ejecuta las celdas de nuevo desde arriba y todo volverá a funcionar.

## 3. Slicing de listas

El **slicing** permite obtener una "rebanada" (un sub-conjunto) de una lista usando la sintaxis `lista[inicio:fin]`. El elemento `inicio` se incluye, `fin` no.

| Sintaxis | Significado |
|----------|-------------|
| `lista[1:4]` | Del índice 1 al 3 (sin incluir el 4) |
| `lista[:3]` | Desde el principio hasta el índice 2 |
| `lista[2:]` | Desde el índice 2 hasta el final |
| `lista[-3:]` | Los últimos 3 elementos |
| `lista[::-1]` | Toda la lista al revés |

Esto es **la base mental para entender la indexación de NumPy y Pandas**, donde más adelante vas a ver cosas como `X_train[:100]` o `df[:, 0]`.

````python
numeros = [10, 20, 30, 40, 50, 60, 70]

print(numeros[1:4])      # del índice 1 al 3
print(numeros[:3])       # primeros 3
print(numeros[3:])       # del índice 3 en adelante
print(numeros[-2:])      # últimos 2
print(numeros[::-1])     # invertida
````

## 4. List comprehensions

Una **list comprehension** es una forma compacta de crear una lista a partir de otra. Sintaxis:

```python
[expresión for elemento in iterable]
```

También se le puede agregar una condición:

```python
[expresión for elemento in iterable if condición]
```

Es **omnipresente** en código de ML: para preprocesar texto, normalizar datos, filtrar columnas, etc.

````python
# Ejemplo 1: doblar cada número
numeros = [1, 2, 3, 4, 5]
dobles = [n * 2 for n in numeros]
print(dobles)

# Ejemplo 2: convertir todos los textos a minúsculas
documentos = ["Hola Mundo", "Python ROCKS", "Machine Learning"]
limpios = [doc.lower() for doc in documentos]
print(limpios)

# Ejemplo 3: con condición — solo los pares
pares = [n for n in numeros if n % 2 == 0]
print(pares)
````

## 5. `enumerate` y `zip`

Dos funciones muy útiles al recorrer listas:

- **`enumerate(lista)`** — al iterar, te da el **índice** y el elemento juntos.
- **`zip(lista1, lista2)`** — al iterar, **empareja** elementos de dos listas, posición por posición.

Ambas son comunes en ML: `enumerate` aparece en loops de entrenamiento (mostrar el número de época), y `zip` para emparejar features con labels.

````python
# enumerate: índice + elemento
frutas = ["manzana", "plátano", "naranja"]
for i, fruta in enumerate(frutas):
    print(i, fruta)

print()

# zip: emparejar dos listas
nombres = ["Ana", "Beto", "Carla"]
notas = [85, 70, 92]
for nombre, nota in zip(nombres, notas):
    print(nombre, "obtuvo", nota)
````

## 1. Tuplas y desempaquetado

Una **tupla** es como una lista, pero **inmutable** (no se puede modificar después de creada). Se escribe con paréntesis: `(1, 2, 3)`.

Lo más útil de las tuplas es que se pueden **desempaquetar** en variables separadas en una sola línea:

```python
x, y = (10, 20)
```

Esto es muy común en ML: muchas funciones devuelven varios valores como tupla. Por ejemplo, `train_test_split` de scikit-learn devuelve 4 elementos:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

````python
# Crear una tupla
punto = (3, 4)
print(punto)

# Desempaquetar en variables separadas
x, y = punto
print("x =", x, " y =", y)

# Una función puede devolver varios valores como tupla
def estadisticas(numeros):
    return min(numeros), max(numeros), sum(numeros) / len(numeros)

minimo, maximo, promedio = estadisticas([10, 20, 30, 40, 50])
print(minimo, maximo, promedio)
````

## 2. Lambda functions

Una **lambda** es una función **anónima** y de **una sola línea**. Sintaxis:

```python
lambda parametros: expresion
```

Equivale a una función `def` corta. Se usan cuando necesitamos pasar una "mini-función" como argumento a otra función. Aparecen muy seguido en código de Pandas (`df.apply(lambda x: ...)`) y en ordenamientos personalizados.

````python
# Definición clásica
def cuadrado(x):
    return x ** 2

# Equivalente con lambda
cuadrado_lambda = lambda x: x ** 2

print(cuadrado(5))
print(cuadrado_lambda(5))

# Uso típico: ordenar por un criterio personalizado
personas = [("Ana", 30), ("Beto", 25), ("Carla", 35)]
ordenadas = sorted(personas, key=lambda p: p[1])  # ordenar por edad
print(ordenadas)
````

## 3. Clases (objetos) — lectura básica

Una **clase** es un molde para crear **objetos**. Un objeto guarda datos (atributos) y funciones (métodos) juntos.

No vamos a aprender a escribir clases complejas — el objetivo es que **entiendas el patrón** que usan todas las librerías de ML:

```python
modelo = LinearRegression()              # crear un objeto
modelo.fit(X_train, y_train)             # llamar a un método (entrenar)
predicciones = modelo.predict(X_test)    # otro método (predecir)
```

Cuando llamamos a `modelo.fit()`, el objeto **guarda internamente** lo que aprendió. Por eso después podemos llamar a `modelo.predict()` y "recuerda" el entrenamiento.

````python
# Definir una clase simple (no te preocupes por la sintaxis exacta)
class Saludador:
    def __init__(self, nombre):
        self.nombre = nombre   # guardar un atributo

    def saludar(self):
        return f"Hola, soy {self.nombre}"

# Crear un objeto a partir de la clase
s = Saludador("Ana")
print(s.nombre)        # acceder a un atributo
print(s.saludar())     # llamar a un método
````

## 4. `*args` y `**kwargs`

`*args` y `**kwargs` permiten a una función aceptar **una cantidad variable de argumentos**:

- `*args` recibe argumentos posicionales extra como una **tupla**.
- `**kwargs` recibe argumentos con nombre extra como un **diccionario**.

No es necesario que escribas funciones así — el objetivo es que las **reconozcas** al leer documentación de librerías. Por ejemplo, vas a ver firmas como:

```python
def fit(self, X, y, **kwargs):
    ...
```

Lo que significa: "esta función acepta `X`, `y`, y cualquier cantidad de argumentos extra con nombre".

````python
# *args: cantidad variable de argumentos posicionales
def sumar_todos(*numeros):
    return sum(numeros)

print(sumar_todos(1, 2, 3))
print(sumar_todos(10, 20, 30, 40, 50))

# **kwargs: argumentos con nombre arbitrarios
def describir(**datos):
    for clave, valor in datos.items():
        print(f"{clave}: {valor}")

describir(nombre="Ana", edad=30, ciudad="Santiago")
````
